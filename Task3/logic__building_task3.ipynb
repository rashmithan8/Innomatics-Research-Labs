```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
highest = max(employees.values())
top = [name for name, score in employees.items() if score == highest]

print("Top Performers Eligible for Bonus:", ", ".join(top), f"(Score: {highest})")
```

```python

# Problem 2: Search Query Keyword Analysis


q= "Buy mobile phone buy phone online"
q = q.lower()
import string
for char in string.punctuation:
    q= q.replace(char, "")
words = q.split()
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1
result = {word: count for word, count in frequency.items() if count > 1}

print("\nKeywords searched more than once:")
print(result)
```

```python

# Problem 3: Sensor Data Validation

sensor = [3, 4, 7, 8, 10, 12, 5]
valid = []
for hour, value in enumerate(sensor):
    if value % 2 == 0:  # Check if even
        valid.append((hour, value))
print("\nValid Sensor Readings (Hour, Value):")
print(valid)
```

```python

# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
count = {}
for email in emails:
    domain = email.split("@")[1]
    count[domain] = count.get(domain, 0) + 1
total = len(emails)
print("\nEmail Domain Usage Percentage:")
for domain, count in count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python

# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]
avg_sales = sum(sales) / len(sales)
threshold = avg_sales * 1.30
print("\nSales Spike Days:")
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python

# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_count = {}
for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1
print("\nDuplicate User IDs:")
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
