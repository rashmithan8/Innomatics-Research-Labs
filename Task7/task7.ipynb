```python
#1
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def switch(self, action):
        self.status = action.upper()

    def show_status(self):
        print(f"{self.name} is {self.status}")


light_name = input("Light Name: ")
action = input("Action: ")

light = SmartLight(light_name)
light.switch(action)
light.show_status()
```

```python
#2
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

employee = Employee(name, emp_id, department)
employee.display_id_card()
```

```python
#3
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def save_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


contact_name = input("Contact Name: ")
phone_number = input("Phone Number: ")

contact = Contact(contact_name, phone_number)
contact.save_contact()
```

```python
#4
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")


product_name = input("Product Name: ")
price = input("Price: ")

product = Product(product_name, price)
product.print_tag()
```

```python
#5
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")


movie_name = input("Movie Name: ")
rating = input("Rating: ")

movie = Movie(movie_name, rating)
movie.display()
```

```python
#6
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def show_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")


customer_name = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(customer_name, address)
delivery.show_details()
```
